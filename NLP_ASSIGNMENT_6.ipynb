```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense
from tensorflow.keras.datasets import imdb

# Load dataset
max_features = 20000  # Vocabulary size
maxlen = 100  # Max sequence length
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Preprocessing
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# Define model creation functions
def create_lstm_model():
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=maxlen))
    model.add(LSTM(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_gru_model():
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=maxlen))
    model.add(GRU(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Create models
lstm_model = create_lstm_model()
gru_model = create_gru_model()

# Train models
lstm_history = lstm_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2)
gru_history = gru_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2)

# Evaluate models
lstm_eval = lstm_model.evaluate(x_test, y_test)
gru_eval = gru_model.evaluate(x_test, y_test)

# Results
print(f"LSTM Test Accuracy: {lstm_eval[1]}")
print(f"GRU Test Accuracy: {gru_eval[1]}")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.datasets import imdb
import numpy as np

# Step (a) Load and Preprocess the Data

# Set parameters
max_features = 10000  # Max number of words in the vocabulary
maxlen = 100  # Max length of each sequence (after padding)

# Load the IMDb dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Pad sequences to ensure uniform input length
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# Step (b) Divide Data into Training and Testing Data Set
# This is already handled by the dataset with x_train, x_test, y_train, and y_test.

# Step (c) Build the Gated Recurrent Units (GRU) Model
def build_gru_model(vocab_size, input_length):
    model = Sequential()
    # Embedding layer: Converts word indices into dense vectors of fixed size
    model.add(Embedding(vocab_size, 128, input_length=input_length))
    # GRU layer: Main RNN layer for processing sequences
    model.add(GRU(128, return_sequences=False))
    # Dense layer: Fully connected layer to output predictions
    model.add(Dense(1, activation='sigmoid'))  # Binary output for sentiment classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Build the GRU model
gru_model = build_gru_model(max_features, maxlen)

# Step (d) Train the GRU Model
gru_model.summary()  # Show the model architecture

# Train the model
history = gru_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2)

# Step (e) Text Generation Using the Trained Model
# Generate text from a sequence - in this case, predict sentiment for a given input
def generate_text(model, input_text):
    input_text = pad_sequences([input_text], maxlen=maxlen)
    prediction = model.predict(input_text)
    return "Positive" if prediction > 0.5 else "Negative"

# Let's take a sample from the test set
sample_text = x_test[0]
print("Generated Sentiment for Test Text:", generate_text(gru_model, sample_text))

# Step (f) Evaluate the Model's Accuracy
test_loss, test_accuracy = gru_model.evaluate(x_test, y_test)

print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# To observe model performance, we can also plot accuracy and loss during training
import matplotlib.pyplot as plt

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
